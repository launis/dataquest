```python
def camel_to_snake(name):
    import re
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()
```

```python
new_columns = []

for c in autos.columns:
    new_columns.append(camel_to_snake(c))

autos.columns = new_columns
```

```python
# Create an empty dictionary to store the results
brand_price_mean = {}
brand_mileage_mean = {}

# Create an array of unique countries
brand_companies = autos["brand"].unique()

# Use a for loop to iterate over the countries
for c in brand_companies:
    # Use boolean comparison to select only rows that
    # correspond to a specific brand
    selected_rows = autos[autos["brand"] == c]
    # Calculate the mean average revenue for just those rows

    brand_price_mean[c] = selected_rows["price"].mean()
    brand_mileage_mean[c] = selected_rows["odometer_km"].mean()

price_mean = pd.Series(brand_price_mean)
mileage_mean = pd.Series(brand_mileage_mean)
price_mean = pd.DataFrame(price_mean, columns=['mean_price'])
mileage_mean = pd.DataFrame(mileage_mean, columns=['mean_mileage'])

result = pd.concat([price_mean, mileage_mean], axis=1)
result["new"] = result["mean_price"]/result["mean_mileage"]

```
