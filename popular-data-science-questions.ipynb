```python
questions2019['Tags']=questions2019['Tags'].str.replace('><', ',').str.replace('<','').str.replace('>','')
```

```python
split_data = questions2019['Tags'].str.split(",")
questions2019['Tags2'] = split_data
#data = split_data.to_list()

# expand df.tags into its own dataframe
tags = questions2019['Tags2'].apply(pd.Series)

# rename each variable is tags
tags = tags.rename(columns = lambda x : 'tag_' + str(x))



# join the tags dataframe back to the original dataframe
questions2019=pd.concat([questions2019[:], tags[:]], axis=1)
```

```python
for ts in questions2019['Tags'].str.split(','):
    for t1 in correlation.index:
        for t2 in correlation.columns:
            if (t1 in ts) and (t2 in ts):
                if correlation.loc[t1,t2] == 0:
                    correlation.loc[t1,t2] = 1
                else:
                    correlation.loc[t1,t2] += 1
```
